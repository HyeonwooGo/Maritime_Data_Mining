```python
# Step 1: 파일 업로드
from google.colab import files
uploaded = files.upload()

# Step 2: JSON 파일 처리
import pandas as pd
import json

# custom.json
with open("custom.json", "r", encoding="utf-8") as f:
    data = json.load(f)
df_custom = pd.json_normalize(data)
print("\n custom.json")
print(df_custom.head(), "\n")

# product.json
with open("product.json", "r", encoding="utf-8") as f:
    data = json.load(f)
df_product = pd.DataFrame(data)
print("product.json")
print(df_product.head(), "\n")

# stock.json
with open("stock.json", "r", encoding="utf-8") as f:
    data = json.load(f)
df_stock = pd.DataFrame(data)
print("stock.json")
print(df_stock.head(), "\n")

# Step 3: XML 파일 처리
import xml.etree.ElementTree as ET

# book.xml
tree = ET.parse("book.xml")
root = tree.getroot()
book_list = []

for book in root.findall("book"):
    book_info = {
        "title": book.find("title").text,
        "author": book.find("author").text,
        "price": float(book.find("price").text),
        "genre": book.find("genre").text
    }
    book_list.append(book_info)

df_book = pd.DataFrame(book_list)
print("book.xml")
print(df_book.head(), "\n")

# news.xml
tree = ET.parse("news.xml")
root = tree.getroot()
news_list = []

for item in root.find("channel").findall("item"):
    news_info = {
        "title": item.find("title").text,
        "link": item.find("link").text,
        "description": item.find("description").text,
        "pubDate": item.find("pubDate").text
    }
    news_list.append(news_info)

df_news = pd.DataFrame(news_list)
print("news.xml")
print(df_news.head(), "\n")

# weather.xml
tree = ET.parse("weather.xml")
root = tree.getroot()

# 현재 날씨
current = root.find("current")
current_info = {
    "city": current.find("city").text,
    "temperature": float(current.find("temperature").text),
    "humidity": int(current.find("humidity").text),
    "condition": current.find("condition").text,
    "wind_direction": current.find("wind").find("direction").text,
    "wind_speed": float(current.find("wind").find("speed").text)
}
df_weather_current = pd.DataFrame([current_info])
print("weather.xml - 현재 날씨")
print(df_weather_current, "\n")

# 예보 정보
forecast_list = []
for day in root.find("forecast").findall("day"):
    temp = day.find("temperature")
    forecast_info = {
        "date": day.attrib["date"],
        "min_temp": float(temp.attrib["min"]),
        "max_temp": float(temp.attrib["max"]),
        "condition": day.find("condition").text,
        "precipitation": float(day.find("precipitation").text)
    }
    forecast_list.append(forecast_info)

df_weather_forecast = pd.DataFrame(forecast_list)
print("weather.xml - 예보")
print(df_weather_forecast.head(), "\n")
```
